```python
import autogen
dir(autogen)
import yfinance as yf
import matplotlib.pyplot as plt
import yfinance as yf
import matplotlib.pyplot as plt
```

Fetching stock price

```python
def sp(ticker):
    try:
        stock = yf.Ticker(ticker)
        history = stock.history(period="1d")

        if history.empty:
            return "No stock data found."

        price = history["Close"].iloc[-1]
        return f"Current price of {ticker.upper()}: ${price:.2f}"
    except Exception as e:
        return f"Error fetching stock data: {str(e)}"
```

After fetching the stock trend, plotted a graph

```python
def st(ticker):
    try:
        stock = yf.Ticker(ticker)
        history = stock.history(period="7d")

        if history.empty:
            return "No data found for this stock symbol."

        plt.figure(figsize=(8, 4))
        plt.plot(history.index, history["Close"], marker="o", linestyle="-", color="blue", label="Closing Price")
        plt.xlabel("Date")
        plt.ylabel("Price (USD)")
        plt.title(f"{ticker.upper()} Stock Trend (Last 7 Days)")
        plt.legend()
        plt.grid()
        plt.xticks(rotation=45)
        plt.show()
        return f"This was the stock trend for {ticker.upper()} "
    except Exception as e:
        return f"Error fetching stock trend: {str(e)}"
```

Built a simple agent

```python
def sa():
    ticker = input("Enter the stock symbol: ").strip().upper()
    command = input(f"Would you like to see the stock price or trend for {ticker}? (Enter 'price' or 'trend'): ").strip().lower()

    if command == "price":
        return sp(ticker)
    elif command == "trend":
        return st(ticker)
    else:
        return "Invalid command. Please enter 'price' or 'trend'."
print(sa())
```
